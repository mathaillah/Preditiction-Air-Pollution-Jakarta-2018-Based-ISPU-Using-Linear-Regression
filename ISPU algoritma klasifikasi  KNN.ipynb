```python
#Import library

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('index_pencermaran_udara_jakarta_2018.csv')
```

```python

df
```

```python
#Melakukan prepocessing data menggunakan mapping dictonary agar dpaat digunakan pada library scikit-learn
critical_map = {'PM10': 1,
                'SO2': 2,
                'O3': 4,
                'CO': 3,
                'NO2': 5
                }

df['critical'] = df['critical'].map(critical_map)

categori_map = {'SANGAT TIDAK SEHAT': 1,
                'TIDAK SEHAT': 2,
                'SEDANG': 3,
                'BAIK': 4
                }
df['categori'] = df['categori'].map(categori_map)

lokasi_spku_map = {'DKI1': 1,
                   'DKI2': 2,
                   'DKI3': 3,
                   'DKI4': 4,
                   'DKI5': 5
                   }
df['lokasi_spku'] = df['lokasi_spku'].map(lokasi_spku_map)
```

```python
# Pisahkan fitur dan target
y = df['categori']
X = df.drop(columns=['categori', 'tanggal',	'max', 'periode_data','critical' ]) #critical, tanggal di drop karena menrupkan data kategorikal tidak bisa dibuat model
```

```python
df
```

```python
df.isnull().sum()
```

```python

# Split data train-test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train model KNN
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
# Evaluasi model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix')
plt.show()
```

```python
Simpan model ke file
with open('project_model_knn.pkl', 'wb') as f:
    pickle.dump(knn, f)

files.download('project_model_knn.pkl')
```
